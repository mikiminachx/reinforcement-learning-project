```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import clone_model
from tensorflow.keras.optimizers import Adam
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
import multiprocessing
import json
from collections import deque
import random
import time


# Function to train the agent and save outputs
df = pd.read_excel('EmotionsIGTData.xlsx', engine='openpyxl')
print(df.head())
```

```python
# Scale IGT scores and Total Money to the range [0, 1]
scaler = MinMaxScaler()
df[['IGT_NET_Raw', 'IGT_NET_1Raw', 'IGT_NET_2Raw', 'IGT_NET_3Raw', 'IGT_NET_4Raw', 'IGT_NET_5Raw', 'Deck_A_Raw', 'Deck_B_Raw', 'Deck_C_Raw', 'Deck_D_Raw', 'Total_Money']] = scaler.fit_transform(df[['IGT_NET_Raw', 'IGT_NET_1Raw', 'IGT_NET_2Raw', 'IGT_NET_3Raw', 'IGT_NET_4Raw', 'IGT_NET_5Raw', 'Deck_A_Raw', 'Deck_B_Raw', 'Deck_C_Raw', 'Deck_D_Raw', 'Total_Money']])
```

```python
class CustomIGTEnvironment:
    def __init__(self, data, noise_factor=0.1):
        self.data = data
        self.num_trials = len(data)
        self.current_trial = 0
        self.noise_factor = noise_factor

    def reset(self):
        self.current_trial = 0
        initial_state = np.array([0, 0, 0, 0])  # Initial state, assuming all decks are unchosen
        return initial_state

    def step(self, action):
        if self.current_trial >= self.num_trials:
            raise ValueError("Episode is already done. Call reset() to start a new episode.")

        current_trial_data = self.data.iloc[self.current_trial]
        self.current_trial += 1
        state_space = np.array([current_trial_data['IGT_NET_Raw'], current_trial_data['IGT_NET_1Raw'], current_trial_data['IGT_NET_2Raw'], current_trial_data['IGT_NET_3Raw']])
         
        reward = self.calculate_reward(state_space, action)
        reward += np.random.normal(0, self.noise_factor)

        done = (self.current_trial >= self.num_trials)

        return state_space, reward, done

    def calculate_reward(self, state, action):
        chosen_deck = action
        reward = 0  # Initialize the reward

        # Define reward rules based on the IGT outcomes
        if chosen_deck == 0:
            # Deck A: Always yields $50
            reward = 50
            # Apply frequent large negative consequences ("punishments")
            if np.random.rand() < 0.5:
                reward -= 250
        elif chosen_deck == 1:
            # Deck B: Always yields $50
            reward = 50
        elif chosen_deck == 2:
            # Deck C: Always yields $100
            reward = 100
            # Apply frequent large negative consequences ("punishments")
            if np.random.rand() < 0.5:
                reward -= 250
        elif chosen_deck == 3:
            # Deck D: Always yields $100
            reward = 100

        return reward
```

```python
# Define the CustomRLAgent class
class CustomRLAgent:
    def __init__(self, state_dim, action_dim, learning_rate=0.001, discount_factor=0.99, epsilon=0.1, param1=0.001, param2=0.1):
        self.learning_rate = learning_rate
        self.optimizer = tf.keras.optimizers.Adam(learning_rate=learning_rate)
        self.model = self.build_model(state_dim, action_dim)
        self.target_model = self.build_model(state_dim, action_dim)
        self.epsilon = epsilon
        self.discount_factor = discount_factor
        self.action_dim = action_dim
        self.replay_buffer = deque(maxlen=10000)  # Experience replay buffer
        self.param1 = param1  # You need to add this parameter

    def build_model(self, state_dim, action_dim):
        model = tf.keras.Sequential([
            tf.keras.layers.Dense(64, activation='relu', input_shape=(state_dim,)),
            tf.keras.layers.Dense(64, activation='relu'),
            tf.keras.layers.Dense(action_dim, activation='linear')
        ])
        model.compile(loss='mse', optimizer=self.optimizer)
        return model

    def select_action(self, state):
        state = np.array(state)
        state = state.reshape(1, -1)
        q_values = self.model.predict(state)

        if np.random.random() < self.epsilon:
            return np.random.choice(self.action_dim)
        else:
            return np.argmax(q_values[0])

    def train(self, state, action, reward, next_state, done):
        next_state = np.array(next_state).reshape(1, -1)
        target_q_values = self.target_model.predict(next_state)
        self.replay_buffer.append((state, action, reward, next_state, done))

        if done:
            target_q_values[0][action] = reward
        else:
            target_q_values[0][action] = reward + self.discount_factor * np.max(target_q_values[0])

        state = np.array(state).reshape(1, -1)

        with tf.GradientTape() as tape:
            q_values = self.model(state)
            loss = tf.keras.losses.mse(q_values, target_q_values)

        gradients = tape.gradient(loss, self.model.trainable_variables)
        self.optimizer.apply_gradients(zip(gradients, self.model.trainable_variables))

        if self.epsilon > 0.01:
            self.epsilon *= 0.9999

    def experience_replay(self, batch_size):
        if len(self.replay_buffer) < batch_size:
            return

        minibatch = np.array(random.sample(self.replay_buffer, batch_size))
        states = np.vstack(minibatch[:, 0])
        actions = minibatch[:, 1]
        rewards = minibatch[:, 2]
        next_states = np.vstack(minibatch[:, 3])
        dones = minibatch[:, 4]

        target_values = self.model.predict(states)
        next_target_values = self.target_model.predict(next_states)

        for i in range(batch_size):
            if dones[i]:
                target_values[i][int(actions[i])] = rewards[i]
            else:
                target_values[i][int(actions[i])] = rewards[i] + self.discount_factor * np.max(next_target_values[i])

        self.model.fit(states, target_values, epochs=1, verbose=0)

    def update_target_model(self):
        self.target_model.set_weights(self.model.get_weights())
```

```python
max_training_time = 3600  # Set the maximum training time to 1 hour (in seconds)

# Function to train the agent and save outputs with hyperparameter tuning, early stopping, and experience replay
def train_agent(param1, param2, model_save_path, num_episodes=50, max_steps=200, batch_size=32, target_reward=100):
    # Create the environment
    env = CustomIGTEnvironment(df)

    # Create the agent with the correct state_dim and action_dim
    state_dim, action_dim = get_state_dim_and_action_dim(env)
    agent = CustomRLAgent(state_dim=state_dim, action_dim=action_dim, param1=param1, param2=param2)

    episode_rewards = []
    training_history = []
    start_time = time.time()  # Record the start time of training

    for episode in range(num_episodes):
        state = env.reset()
        episode_reward = 0

        for step in range(max_steps):
            action = agent.select_action(state)
            next_state, reward, done = env.step(action)

            episode_reward += reward

            agent.train(state, action, reward, next_state, done)

            if done:
                break

            state = next_state

        episode_rewards.append(episode_reward)

        # Update the target model
        if episode % 10 == 0:
            agent.update_target_model()

        # Experience replay
        agent.experience_replay(batch_size)

        # Early stopping if the target reward is achieved or time limit is reached
        if np.mean(episode_rewards[-10:]) >= target_reward or time.time() - start_time >= max_training_time:
            print(f"Training stopped after episode {episode + 1} with mean reward {np.mean(episode_rewards[-10:])}")
            break

        training_history.append({
            'episode': episode,
            'average_reward': np.mean(episode_rewards)
        })

    # Save the trained model with a unique name
    agent.model.save(model_save_path)

    return param1, param2, np.mean(episode_rewards), training_history

# Get the state_dim and action_dim from the environment
def get_state_dim_and_action_dim(env):
    state_dim = 4  # Assuming your state has 4 dimensions
    action_dim = 4  # Assuming your action space has 4 choices
    return state_dim, action_dim

if __name__ == '__main__':
    # Define the hyperparameters for multiprocessing
    num_processes = 6  # Number of parallel processes
    param1_values = [0.001, 0.01, 0.1]
    param2_values = [0.1, 0.2, 0.3]

    # Create parameter combinations for multiprocessing
    params = [(param1, param2, f'model_{param1}_{param2}.keras') for param1 in param1_values for param2 in param2_values]

    with multiprocessing.Pool(processes=num_processes) as pool:
        results = pool.starmap(train_agent, params)

    # Calculate and print average rewards for different hyperparameters
    for i, (param1, param2, avg_reward, _) in enumerate(results):
        print(f"Hyperparameters: param1={param1}, param2={param2}, Average Reward: {avg_reward}")

    # Save training history to a JSON file
    with open('training_history.json', 'w') as history_file:
        json.dump(results[0][3], history_file)  # Saving training history from the first result

    # Save the outputs to a text file
    with open('outputs.txt', 'w') as file:
        for param1, param2, avg_reward, _ in results:
            file.write(f"Hyperparameters: param1={param1}, param2={param2}, Average Reward: {avg_reward}\n")

    print("Training is done.")
```
